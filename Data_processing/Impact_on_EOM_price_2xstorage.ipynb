```python
import pandas as pd
#data input
EOM_prices_w_opt = pd.read_csv('/Users/kanankhasmammadov/Desktop/Thesis - Electrolyzer market participation/flexABLE_w_electrolyzer/output/2016_w_opt_2xstorage/EOM_Prices.csv')
EOM_prices_wo_opt = pd.read_csv('/Users/kanankhasmammadov/Desktop/Thesis - Electrolyzer market participation/flexABLE_w_electrolyzer/output/2016_wo_opt/EOM_Prices.csv')

display(EOM_prices_w_opt)
display(EOM_prices_wo_opt)
```

```python
import matplotlib.pyplot as plt

# Calculate the standard deviation for EOM_prices_w_opt
std_w_opt = EOM_prices_w_opt['Price'].std()

# Calculate the standard deviation for EOM_prices_wo_opt
std_wo_opt = EOM_prices_wo_opt['Price'].std()

# Plot the standard deviations
plt.bar(['EOM_prices_w_opt', 'EOM_prices_wo_opt'], [std_w_opt, std_wo_opt])
plt.xlabel('Dataframe')
plt.ylabel('Standard Deviation')
plt.title('Volatility Comparison')
plt.show()

```

```python
# Plot histogram for EOM_prices_w_opt with increased resolution
plt.hist(EOM_prices_w_opt['Price'], bins=80, alpha=0.5, label='EOM_prices_w_opt')

# Plot histogram for EOM_prices_wo_opt with increased resolution
plt.hist(EOM_prices_wo_opt['Price'], bins=80, alpha=0.5, label='EOM_prices_wo_opt')

plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram Comparison')
plt.legend()
plt.show()

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform time series analysis on EOM_prices_w_opt
decomposition_w_opt = seasonal_decompose(EOM_prices_w_opt['Price'], period=12)

# Perform time series analysis on EOM_prices_wo_opt
decomposition_wo_opt = seasonal_decompose(EOM_prices_wo_opt['Price'], period=12)

# Plot the seasonality patterns
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(decomposition_w_opt.trend)
axes[0, 0].set_title('Trend - EOM_prices_w_opt')

axes[0, 1].plot(decomposition_w_opt.seasonal)
axes[0, 1].set_title('Seasonal - EOM_prices_w_opt')

axes[1, 0].plot(decomposition_wo_opt.trend)
axes[1, 0].set_title('Trend - EOM_prices_wo_opt')

axes[1, 1].plot(decomposition_wo_opt.seasonal)
axes[1, 1].set_title('Seasonal - EOM_prices_wo_opt')

plt.tight_layout()
plt.show()
```
